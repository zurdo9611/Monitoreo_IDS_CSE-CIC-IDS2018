```python
import pandas as pd
import numpy as np
import glob
import os

# Colocar la ruta de los archivos CSV
ruta = r"C:\Users\Usuario\OneDrive\Documentos\Maestría\Redes\Proyecto final - Lab\CSE-CIC-IDS2018-Attacks"   

# Buscar todos los CSV
archivos = glob.glob(os.path.join(ruta, "*.csv"))

# Leerlos en una lista de DataFrames
archivos_leidos = []

for archivo in archivos:
    archivo_leido = pd.read_csv(archivo)
    archivos_leidos.append(archivo_leido)

df = pd.concat(archivos_leidos, ignore_index=True)

# Liberar memoria eliminando la lista de DataFrames individuales
del archivos_leidos
```

```python
traffic_types = [
    "Benign",
    "DDOS attack-HOIC",
    "DDoS attacks-LOIC-HTTP",
    "DoS attacks-Hulk",
    "Bot",
    "FTP-BruteForce",
    "SSH-Bruteforce",
    "Infilteration",
    "DoS attacks-SlowHTTPTest",
    "DoS attacks-GoldenEye",
    "DoS attacks-Slowloris",
    "DDOS attack-LOIC-UDP",
    "Brute Force -Web",
    "Brute Force -XSS",
    "SQL Injection"
]

# Lista de ataques (excluyendo Benign)
attack_types = traffic_types[1:]

# 1. Sample de 500 benign
benign_sample = df[df["Label"] == "Benign"].sample(500, random_state=42)

# 2. Sample de ataques distribuidos
attack_samples = []

for attack in attack_types:
    subset = df[df["Label"] == attack]
    
    # Si hay menos de 35, toma todo; si hay más, toma 35
    n = min(35, len(subset))
    attack_samples.append(subset.sample(n, random_state=42))

# Concatenar todos los ataques
attack_sample = pd.concat(attack_samples, ignore_index=True)

# 3. Unir benign + ataques
balanced_df = pd.concat([benign_sample, attack_sample], ignore_index=True)

balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

print(balanced_df["Label"].value_counts())
```

```python

# Guardar en la ruta esperada por tu proyecto
balanced_df.to_csv('sample_traffic.csv', index=False)

print("¡Archivo 'sample_traffic.csv' generado con éxito con 1000 flujos de red mixtos!")
```
